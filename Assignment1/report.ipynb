In Dynamic Programming, the value Iteration algorithm iteratively applies Bellman backup equation to evaluate the value function, e.g., for state-value function:

$$ V_k(s) = \max{a \in A}\left(r + \gamma \sum_{s' \in S} P(s'|s,a) V_{k-1}(s') \right) $$

Write the bellman backup equation for action-value function (Q-function)
</span>



Bellman backup equation for action value in dynamic programming is given by
The question assumes a deterministic reward. Considering that

$$ Q_k(s, a) = r + \gamma \sum_{s' \in S} P(s'|s,a) \max_{a'} Q_{k-1}(s', a') $$


Where, <br>
<pre>
    k is the iteration number <br>
    r is the reward (deterministic reward) <br>
    Q is the Q value for each state s and action a pair <br>
    s is the state <br>
    a is the action <br>
    s' is the next state <br>
    a' is the next action <br>
    P(s'|s,a) is the probability of next state s' given current state,action pair

## Comparision between using TD(0) vs Q-learning

1. Looking at the Total Rewards in episodes <br>
    Using V(s): The total rewards varies a lot over the episodes. This learning method is quick to find the positive rewards here but there are multiple large dips (high variance) in the graph even at later episodes indicating a bit of instability in learning. So it hard to say when there is convergence without a clear trend. <br>
    Using Q(s,a): On the other hand, Q-learning has a reward curve which smoothly increases over the episodes. This method find the positive rewards comparitively slower to TD(0) state value method but it starts from a low value and gradually improves with episodes and shows a clear convergence. The clear convergence shows a more stable learning. 
    <br>
    <br>
2. Looking at the Total steps per episodes <br>
    Using V(s): Similar to total rewards graph, the total steps per episodes graph shows the number of steps per episode remains inconsistent. There is no clear decreasing trend in the grpah. This shows a high variance in number of steps. <br>
    Using Q(s,a): Initially the agent takes a very large number of steps exploring the env with various state and action pair. After a certain episodes, the number of steps decreases and stabilizes at a lower value. This shows a good learning curve.
    <br>
    <br>
3. Looking the max V(s) vs max Q(s,a) <br>
    Using V(s): This method shows a distinction in max V(s) values with less number of episodes. <br>
    Using Q(s,a): Compared to the other method, Q-learning takes more episode to stabilize the values of Q. But the gradient is more distinct once there is a stabilization. This again asserts the point that Q-learning is more stable.
    <br>
    <br>
4. Policy: 
    Finding optimal policy using Q is easier and more accurate. Q directly tells us about both state and action which leads to maximize rewards. On the other hand, when using V(s) we need to check all possible actions and then get the policy.
    <br>
    <br>
5. Changing the parameters 
    Learning rate: Changing learning rate from 0.01 to 0.02, changes the rate of convergence in both the cases. The TD(0) value function case still has high variance in both total rewards per episode as well as total steps per episode. In case of Q learning, it results in a faster convergence but both rewards per episode and steps per episode are stable.

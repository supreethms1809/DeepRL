In Dynamic Programming, the value Iteration algorithm iteratively applies Bellman backup equation to evaluate the value function, e.g., for state-value function:

$$ V_k(s) = \max{a \in A}\left(r + \gamma \sum_{s' \in S} P(s'|s,a) V_{k-1}(s') \right) $$

Write the bellman backup equation for action-value function (Q-function)

Bellman backup equation for value iteration in dynamic programming is given by
The question assumes a deterministic reward. Considering that

**Deterministic reward**

$$ Q_k(s, a) = r + \gamma \sum_{s' \in S} P(s'|s,a) \max_{a'} Q_{k-1}(s', a') $$


Where, <br>
<pre>
    k is the iteration number <br>
    Q is the Q value for each state s and action a pair <br>
    s is the state <br>
    a is the action <br>
    s' is the next state <br>
    a' is the next action <br>
    P(s'|s,a) is the probability of next state s' given current state,action pair
